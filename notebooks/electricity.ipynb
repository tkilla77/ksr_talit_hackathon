```python
import pandas as pd

electricity_price = pd.read_table('https://bfe-energy-dashboard-ogd.s3.amazonaws.com/ogd106_preise_strom_boerse.csv', sep=',')
```

```python
electricity_price.Datum = pd.to_datetime(electricity_price.Datum)
electricity_price.set_index('Datum', inplace=True)
electricity_price
```

```python
import pandas as pd
import plotly

for period in ['W', 'M', 'Q']:
    electricity_price[period] = electricity_price.index.to_period(period)
    electricity_price.reset_index().groupby(period).mean(numeric_only=True).plot(subplots=True, title="Price per " + period)
```
